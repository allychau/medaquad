```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import csv
```

```python
covid_data_file1 = '../data/race_data_tracker.csv'
covid_tracker_df = pd.read_csv(covid_data_file1, dtype="object", encoding="utf-8")
# covid_tracker_df.shape
covid_tracker_df.tail(50)
```

```python
top_states_df = covid_tracker_df[(covid_tracker_df['State'] == 'CA') & (covid_tracker_df['Date'] == '20200722') |
                          (covid_tracker_df['State'] == 'FL') & (covid_tracker_df['Date'] == '20200722') |\
                          (covid_tracker_df['State'] == 'NY') & (covid_tracker_df['Date'] == '20200722')|\
                          (covid_tracker_df['State'] == 'TX') & (covid_tracker_df['Date'] == '20200722')]
top_states_df
```

```python
top_states_df.columns
```

```python
# Cases
state_cases_df = top_states_df[['State','Cases_Total','Cases_White', 'Cases_Black',
       'Cases_LatinX', 'Cases_Asian', 'Cases_AIAN', 'Cases_NHPI']]
state_cases_df.head()
```

```python
# Declare a list that is to be converted into a column 
population = ['39512223', '21477737', '19453561', '28995881'] 

# Using 'Address' as the column name 
# and equating it to the list 
state_cases_df['Population'] = population

state_cases_df
```

```python
top_states_df.columns
```

```python
#Deaths
state_deaths_df= top_states_df[['State','Deaths_Total', 'Deaths_White', 'Deaths_Black', 'Deaths_LatinX',
       'Deaths_Asian', 'Deaths_AIAN', 'Deaths_NHPI']]
state_deaths_df
```

```python
# Export the clean covid tracker data into a csv
state_cases_df.to_csv("../data/state_cases.csv",encoding='utf-8', index=False)
```
