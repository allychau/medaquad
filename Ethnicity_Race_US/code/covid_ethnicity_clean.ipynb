```python
pop_data_file = '../data/est_ethnicity_pop2017_census.csv'
pop_data_df = pd.read_csv(pop_data_file, dtype="object", encoding="utf-8")
pop_data_df.head()
```

```python
pop_data_df.columns
```

```python
# Keep only needed columns
pop_data_df = pop_data_df[['State','Origin','Race','Pop_Estimate2017']]
pop_data_df.head(5)
```

```python
#Replace column ORIGIN Value from 0 = Total,1 = Not Hispanic, 2 = Hispanic
pop_data_df['Origin'] = pop_data_df['Origin'].replace(['0','1','2'],['Total','Not Hispanic', 'Hispanic'])
```

```python
pop_data_df.head()
```

```python
pop_data_df.tail()
```

```python
# Replace column RACE values: 
# 1 = White Alone
# 2 = Black or African American Alone
# 3 = American Indian and Alaska Native Alone
# 4 = Asian Alone
# 5 = Native Hawaiian and Other Pacific Islander Alone
# 6 = Two or more races

pop_data_df['Race'] = pop_data_df['Race'].replace(['1','2','3','4','5','6'],
                                                  ['White Alone',
                                                  'Black or African American Alone',
                                                  'American Indian and Alaska Native Alone',
                                                  'Asian Alone',
                                                  'Native Hawaiian and Other Pacific Islander Alone',
                                                  'Two or more races'])
```

```python
pop_data_df.head(10)
```

```python
pop_data_df.tail(10)
```
